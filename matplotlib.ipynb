```python
x = [1,2,3,4]
y = [2,3,5,10]

plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

plt.fill_between(x[2:4], y[2:4], alpha = 0.5)

plt.show()
```

```python
x = [1,2,3,4]
y1 = [2,3,5,10]
y2 = [1,2,4,8]

plt.plot(x,y1)
plt.plot(x,y2)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.fill_between(x[1:3], y1[1:3], y2[1:3], color='gray', alpha=0.5)

plt.show()
```

```python
import pandas as pd
titanic_df = pd.read_csv('titanic_train.csv')
```

```python
titanic_df2 = titanic_df.loc[:,['Age','Fare']]
```

```python
titanic = titanic_df2.to_numpy()
```

```python
titanic
```

```python
x= titanic[:, [0]]
y= titanic[:, [1]]
```

```python
print(x)
print(y)
```

```python
plt.plot(x,y,'k')
plt.show()
```

```python
fig, axs = plt.subplots(nrows = 1, ncols = 2, figsize=(12, 6))

axs[0].scatter(x, y, marker = '*', color = 'k')
axs[0].set_title('Scatter Plot Black Stars')
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')

axs[1].scatter(x, y, marker = 'o', color = 'r')
axs[1].set_title('Scatter Plot Red Circles')
axs[1].set_xlabel('x')
axs[1].set_ylabel('y')

plt.show()
```
